```python
# Define the number of characters per line
characters_per_line = 150

# Open the file in write mode
with open("output.txt", "w") as file:
    # Loop through the string and write chunks of the specified length
    for i in range(0, len(content), characters_per_line):
        # Write a chunk of the string (from i to i + characters_per_line)
        file.write(content[i:i + characters_per_line] + "\n")
```
